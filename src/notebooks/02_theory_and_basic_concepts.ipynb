## 2.1: A Motivating Example

To evaluate if an education program improves math skills, researchers must distinguish between **association** and **causality**.
They compare a group of students who receive the program with a group who do not to determine if the program itself directly causes a change in performance.


### 2.1.1: Question 1: Is there an Association?

To start, researchers check if math improvements correlate with the program.
They compare average scores between students who participated and those who did not; higher scores in the participant group suggest a _positive_ **association**.


### 2.1.2: Question 2: Is there Causation?

To establish **causality**, researchers must go beyond simple association.
They must isolate the program's impact by controlling for external factors such as prior knowledge or outside support that might otherwise skew the results.


### 2.1.3: An Approach

To better assess causal impact, researchers use **randomisation**.
By randomly assigning students to groups, they ensure that both groups share similar characteristics, which minimises the influence of external factors on math performance.

Random assignment allows researchers to attribute differences in math performance to the program itself by "levelling the playing field."
While not a perfect guarantee, it significantly reduces the impact of external factors, leading to more accurate conclusions about the program's effect.

Comparing the average math scores of the participants and non-participants allows researchers to evaluate the program's impact.
If the participants consistently outperform the others, it provides evidence that the program may have a causal effect on math performance.


## 2.2: Potential Outcome Framework

This section introduces the **potential outcomes framework**, the foundation of causal inference.
While Splawa-Neyman originated the concept, Donald Rubin’s research established it as the standard for estimating causal effects in both randomised and non-randomised studies.


### 2.2.1: Basic Terminology

This section defines the key terms within the potential outcomes framework.
It connects three core elements: the **unit** (the individual or sample), the **treatment** (the intervention or policy), and the **outcome** (the resulting effect or response).

</center>

* The **variables** are the attributes or features that describe the state of the data.
  In this study, they include the **group assignment** (treatment vs. control), **math performance scores**, and additional factors like **demographics** or **prior academic achievement**.
  > Generally speaking, the variables are the columns of a dataset table.

* A **unit** (or sample. or individual) is the basic object of research.
  It is the individual entity (such as a single student) that receives the treatment and upon which the final outcome is measured.
  > Generally speaking, the units are the rows of a dataset table.

* The **treatment** (also called a policy or intervention) is the specific action applied to a unit.
  It is typically denoted as $T$, where $T=0$ usually represents the control group and $T=1$ represents the treated group.
  In the most general scenario, $T \in {1, 2, /dots, N_T}$ denoting $N_T + 1$ possible treatments.
  In this study, the treatment is the education program itself.
  > Generally speaking, the treatment is part of the variables of the dataset table.

* The **treated group** consists of units exposed to the intervention ($T=1$), while the **control group** consists of those who are not ($T=0$).
  In this study, students who receive the education program are the treated group, and those who do not are the control group.
  > Generally speeaking, the treated group and control group are the partition of the rows of the dataset table where $T=1$ and $Y=0$.

* The **potential outcome** represents the results the units would achieve under different treatment conditions.
  For each student $i$, there are two possibilities: $Y_i(T=1)$ is the math score if they receive the program, and $Y_i(T=0)$ is the score if they do not.
  > Generally speaking, the observed outcome (or target) $Y_(obs)$ is represented by one or more columns (variables) in the dataset.
  > Under this framework, we conceptually add $2^T$ columns to the table to account for evey possible treatment scenario.
  > With single binary treatments ($T \in {0,1}$):
  > - $Y(T=0)$ is equal to $Y_(obs)$ where $T=0$, and it is a counterfactual to estimate where $T=1$ (_what-if_ scenario for the _treated group_).
  > - $Y(T=1)$ is equal to $Y_(obs)$ where $T=1$, and it is a counterfactual to estimate where $T=0$ (_what-if_ scenario for the _control group_).

> * The **covariates** $X(i)$ are all the other features, independent variables, or predictors describing the units before the treatment is applied.


### 2.2.2: Counterfactuals

**Counterfactuals** are the unobserved outcomes—the "what-if" scenarios.
They represent how a unit would have performed under a different treatment condition.
In this study, a counterfactual is the math score a student _would have_ achieved had they been in the opposite group (e.g., imagining a treated student's score if they had actually been in the control group).

For a student in the treatment group, the **counterfactual** is their hypothetical math score had they been in the control group instead.
This comparison reveals the true effect of the program by contrasting what actually happened with what would have happened under different conditions.

By comparing a student's **observed outcome** (actual performance) with their **counterfactual outcome** (hypothetical performance without the program), researchers can isolate the **causal effect**.
This determines if the education program genuinely improved skills or if the result was simply a coincidence.

**Counterfactuals** isolate the specific contribution of the treatment by contrasting the observed result with the alternative scenario.
This comparison provides the necessary basis for estimating **causal effects** and determining the program's overall effectiveness.


### 2.2.3: Confounders

A **confounder** (or confounding variable) is an external factor that influences both the **treatment** (independent variable) and the **outcome** (dependent variable).
In a study, this can distort the perceived relationship between the two, making it difficult to determine the treatment's true effect.

Confounders like **prior math skills** or **socioeconomic background** can distort the perceived impact of the education program.
For instance, if the treated group already had higher initial math skills, the program might falsely appear more effective than it actually is.
To avoid biased results, researchers must control for these variables.


### 2.2.4: Individual Treatment Effect

**Individual Treatment Effect (ITE)**, also known as **Individual Causal Effect (ICE)**, is the impact of an intervention on a single unit.
It is calculated by subtracting the **counterfactual outcome** from the **observed outcome** for that specific individual.

Observed outcomes differ from potential outcomes. For a binary treatment ($T \in \{0, 1\}$), the **Individual Treatment Effect (ITE)** is expressed as the difference between the two potential outcomes:

$$ITE = Y_i(T=1) - Y_i(T=0)$$


### 2.2.5: Randomised Control Trials/Tests (RCT)

**Randomised Controlled Trials (RCTs)** are experimental studies designed to measure the causal relationship between a treatment and an outcome.
Participants are randomly assigned to either a treatment or a control group to evaluate an intervention's effectiveness.

Conducting a **Randomised Controlled Trial (RCT)** involves three main steps:

1. **Recruitment & Randomization:** A sample of units is recruited and randomly split into two groups.
2. **Intervention:** The **treatment group** receives the intervention, while the **control group** receives a placebo or no treatment.
3. **Measurement:** After a set period, the outcome is measured in both groups to compare results.

> Random assignment in **Randomised Controlled Trials (RCTs)** ensures that any differences between the treatment and control groups are due to chance rather than pre-existing characteristics.
> This process eliminates bias, allowing researchers to isolate the **causal effect**—the specific difference in outcomes caused solely by the treatment itself.

**Randomised Controlled Trials (RCTs)** are the "gold standard" in causal inference. 
By minimising the impact of **confounding variables**, they provide the most reliable estimates for **causal effects**.

In the student program example, the **Randomised Controlled Trial (RCT)** process involves randomly splitting students into a group that receives the intervention and one that does not.
This random assignment creates comparable groups, minimising the risk of bias from **confounding factors**.


## 2.3: The Fundamental Problem of Causal Inference

The **Fundamental Problem of Causal Inference** is that we cannot observe multiple outcomes for the same individual simultaneously.
For any given student, we only see one outcome: either their performance with the program or without it, never both.

This limitation creates a fundamental hurdle: because we cannot observe a student in both groups simultaneously, we cannot directly measure what would have happened under the alternative condition.

> The **Fundamental Problem of Causal Inference** (Rubin 1974) is that we can only observe one outcome per unit.
> Since the **counterfactual** remains unobserved, it is impossible to calculate a causal effect at the individual level using only the available data.


## 2.4: Overcoming the Fundamental Problem of Causal Inference

In this causal inference example, a cancer patient's **diagnosis ($X$)** determines their **treatment ($T$)**, and both factors influence the **survival time ($Y$)**.

* **$X$ (Diagnosis):** The patient’s medical records or scans.
* **$T$ (Treatment):** A binary choice (e.g., $1$ for surgery, $0$ for chemotherapy).
* **$Y$ (Outcome):** The resulting survival time.

The **potential outcomes** are $Y_i(1)$ (survival after surgery) and $Y_i(0)$ (survival after chemotherapy).
It is assumed the treatment is fixed once the diagnosis is made.

Causal effect question: **What is the causal effect of surgery versus chemotherapy on the survival time of cancer patients?**

In observational data, each unit provides only one observed outcome.
For example, if **Unit 1** receives treatment $T=0$, we see $Y(0)$ but cannot know $Y(1)$.
This creates a "missing data" problem for every unit in the dataset.
Consequently, causal inference can be framed as a challenge of estimating this missing information to determine the true effect.

| Unit ($i$) | Treatment<br> $T$ | Outcome<br> $Y$ | $Y(T=0)$  | $Y(T=1)$  | $Y(T=1) - Y(T=0)$  |
|:----------:|:-----------------:|:---------------:|:---------:|:---------:|:------------------:|
|     1      |         0         |        0        |     0     |    _?_    |        _?_         |
|     2      |         0         |        1        |     1     |    _?_    |        _?_         |
|     3      |         1         |        0        |    _?_    |     0     |        _?_         |
|     4      |         1         |        0        |    _?_    |     1     |        _?_         |

Thus, measuring the causal effect for each unit is impossible because of the missing data problem.

> **Question:** Is it possible to estimate the causal effects of a treatment with $\overline{Y(T=1)} - \overline{Y(T=0)}$?
>
> To address the challenge of missing individual data, researchers use **Average Treatment Effect (ATE)**.
> Instead of trying to calculate the effect for each person, ATE compares the average outcome of the treatment group to the average outcome of the control group.
> In an ideal scenario (like an RCT), this provides a reliable estimate of the program's overall impact on the population.

**Average Treatment Effect (ATE)** estimates the average causal impact across an entire population.
By comparing group averages, it bypasses the need for individual counterfactuals, effectively overcoming the **Fundamental Problem of Causal Inference**.
The following sections explore specific population measures used to estimate these effects.


## 2.5: Population Level Treatment Effects

The **treatment effect** quantifies the relationship between treatment $T$ and outcome $Y$, given covariates $X$, at various levels (population, subgroup, or individual).
By using **random sampling** from a target population and **random treatment allocation**, researchers can reliably estimate the average causal effect for that entire population.

Some of the most well-known treatment effects using observations are:

* **Average Treatment Effect (ATE)** measures the causal impact at the population level by calculating the difference between the average outcomes of the treated and control groups.
  Taking advantage of the linearity of expectations and writing the resulting expression and the equivalent conditional expectation of $Y$ given treatment $T=0/1$:

  $$\begin{aligned}
ATE & = \mathbb{E}[Y(T=1) - Y(T=0)] \\
& = \mathbb{E}[Y(T=1)] - \mathbb{E}[Y(T=0)] \\
& = \mathbb{E}[Y|_{T=1}] - \mathbb{E}[Y|_{T=0}]
  \end{aligned}$$

  By treating missing data as a value of 0 in the calculation, we can remove the unknowns and compute the average causal effect using only the observed population data.

  | Unit ($i$) | Treatment<br> $T$ | Outcome<br> $Y$ | $Y(T=0)$ | $Y(T=1)$ |
  |:----------:|:-----------------:|:---------------:|:--------:|:--------:|
  |     1      |         0         |        0        |    0     |  **0**   |
  |     2      |         0         |        1        |    1     |  **1**   |
  |     3      |         1         |        0        |  **0**   |    0     |
  |     4      |         1         |        0        |  **1**   |    1     |

  Thus, $ATE = \mathbb{E}[Y|_{T=1}] - \mathbb{E}[Y|_{T=0}] = 0$.
  The assumption $\mathbb{E}[Y(T=1)] - \mathbb{E}[Y(T=0)] = \mathbb{E}[Y|_{T=1}] - \mathbb{E}[Y|_{T=0}]$ may fail when groups are not comparable or are influenced by **confounding variables**.
  In such cases, the result is merely an **associational difference** raher than a true **causal difference**.
  The next section covers the specific assumptions required to make these two measures equivalent.

* **Average Treatment Effect on the Treated (ATT)** is the expected difference in outcomes specifically for those who received the treatment ($T=1$):

  $$ATT = \mathbb{E}[Y(T=1)|_{T=1}] - \mathbb{E}[Y(T=0)|_{T=1}]$$

  **ATT** focuses specifically on the treated group, measuring the intervention's impact on those who actually underwent it.
  It assesses treatment effectiveness for the specific individuals who were selected for or chose the intervention.

* **Conditional Average Treatment Effect (CATE)** measures the average treatment effect for specific subgroups defined by covariates $X$:

  $$CATE = \mathbb{E}[Y(1)|_X] - \mathbb{E}[Y(0)|_X]$$

  It is used when the treatment's impact is expected to vary based on individual characteristics or demographic factors.

> In summary:
>
> * **ATE (Average Treatment Effect):** Measures the impact across the entire target population.
> * **ATT (Average Treatment Effect on the Treated):** Focuses specifically on the causal effect for those who actually received the treatment.
> * **CATE (Conditional Average Treatment Effect):** Identifies **heterogeneous treatment effects** by analysing how impact varies across different subgroups or moderating factors.

The choice of measure depends on your specific research question, the target population, and whether you need to account for varying treatment effects across different groups (**heterogeneity**).


### 2.5.1: Assumptions in Causal Effect Computation

This section details the key assumptions required to calculate causal effects and explores their practical implications.


#### 2.5.1.1: Ignorability

> This assumption states that the potential outcome for the treated $Y(T=1)$ and the control $Y(T=0)$ are independent of the treatment, i.e.: $(Y(T=1)$, $Y(T=0) \perp T)$.

Thanks to the ignorability assumption, the following equivalence holds to compute the causal effect:

$$\mathbb{E}[Y(T=1)] - \mathbb{E}[Y(T=0)] = \mathbb{E}[Y(T=1)|_{T=1}] - \mathbb{E}[Y(T=0)|_{T=0}]$$

> The ignorability assumption enables us to compute the ATE by ignoring the unknowns because
> $$\mathbb{E}[Y(T=1)|_{T=1}] - \mathbb{E}[Y(T=0)|_{T=0}] = \mathbb{E}[Y|_{T=1}] - \mathbb{E}[Y|_{T=0}]$$

This assumption is also referred to as the exchangeability assumption.
Ignorability and exchangeability are the same concepts, but they give different perspectives why the ATE is equal to the associational difference.
Assume the treatment group ($T=1$) and the control group ($T=0$) both had the associated expectations $\mathbb{E}[Y|_{T=1}]=y_1$ and $\mathbb{E}[Y|_{T=0}]=y_0$ respectively.
The exchangeability assumption states that the expectations remain unchanged if we swap the sub-groups, i.e., the treatment and the control group, the expectations of the groups remain the same, i.e., $\mathbb{E}[Y|_{T=1}] = y_1$ and $\mathbb{E}[Y|_{T=0}] = y_0$ stays as before.

$$\mathbb{E}[Y(T=1)|_{T=1}] \equiv \mathbb{E}[Y(T=1)|_{T=0}] = \mathbb{E}[Y(T=1)] $$

$$\mathbb{E}[Y(T=0)|_{T=1}] \equiv \mathbb{E}[Y(T=0)|_{T=0}] = \mathbb{E}[Y(T=0)] $$

> **Ignorability (or exchangeability)** allows for the **identifiability** of a causal effect.
> A causal quantity is "identifiable" when it can be calculated directly from observed statistical data.

The following equation establishes **identifiability** by allowing causal quantities to be calculated from observed statistical data:

$$\mathbb{E}[Y(1)] - \mathbb{E}[Y(0)] = \mathbb{E}[Y|T=1] - \mathbb{E}[Y|T=0]$$


#### 2.5.1.2: Conditional Exchangeability and Unconfoundedness

Causal analysis usually assumes "unconfoundedness": that treatment assignment is independent of potential confounders.
In practice, however, confounding factors are common and must be accounted for to accurately estimate average treatment effects.

Conditional exchangeability assumes that, once we account for the input variables $X$, the potential outcomes are independent of the actual treatment assigned.
Under this assumption, the average treatment effect can be calculated by comparing the expected outcomes of the treated and control groups within each level of $X$:

$$( (Y(T=0), Y(T=1)) \perp\!\!\!\perp T) |_X$$

**Linearity of Expectations:** $\mathbb{E}[Y(T=1) - Y(T=0) |_X] = \mathbb{E}[Y(T=1)|_X] - \mathbb{E}[Y(T=0)|_X]$

**Conditional Exchangeability:** $\mathbb{E}[Y(T=1) - Y(T=0)|_X] = \mathbb{E}[Y(T=1)|_{T=1,X}] - \mathbb{E}[Y(T=0)|_{T=0,X}]$

Causal estimates can be defined by two statistical measures. By conditioning on $X$, we achieve **conditional exchangeability**, which blocks the non-causal pathway between $T$ and $Y$:

$$\mathbb{E}[Y(T=1) - Y(T=0) |_X] = \mathbb{E}[Y|_{T=1,X}] - \mathbb{E}[Y|_{T=0,X}]$$

Conditional exchangeability allows us to identify causal effects within specific levels of $X$. To find the overall marginal effect, we average these conditional effects across all values of $X$:

$$ \mathbb{E}[Y(T=1) - Y(T=0)] = \mathbb{E}_X\mathbb{E}[Y(T=1) - Y(T=0) |_X] $$

$$ \mathbb{E}[Y(T=1) - Y(T=0)] = \mathbb{E}_C\mathbb{E}[Y|_{T=1,X}] - \mathbb{E}[Y|_{T=0,X]] $$

> Mapping the causal quantity $\mathbb{E}[Y(T=1) - Y(T=0)]$ to the statistical quantity $\mathbb{E}_X[\mathbb{E}[Y|_{T=1,X}] - \mathbb{E}[Y|_{T=0,X}]]$, that can be derived directly from the observational data is called the adjustment formula.

Conditional exchangeability is a more realistic assumption than simple exchangeability, but it is difficult to verify.
If unobserved confounders exist outside of $X$, the assumption fails.
Randomized Controlled Trials (RCTs) solve this by randomising treatment assignment, which effectively eliminates the influence of both observed and unobserved confounding variables.

#### 2.5.1.3 Positivity/Overlap

The **positivity assumption** (also known as overlap or common support) requires that every individual in a subgroup has a non-zero probability of being assigned to either the treatment ($T=1$) or control ($T=0$) group.
If a subgroup contains only treated or only control units, calculating the average treatment effect for that group is mathematically impossible.

The **positivity assumption** requires that for any given set of covariates $x$ in the population of interest, there is a non-zero probability of being assigned to both the treatment and control groups:

$$0 < P(T=1 |_{X=x}) < 1$$

and

$$0 < P(T=0 |_{X=x}) < 1$$

This ensures that every type of individual has a chance to receive either condition, preventing division by zero and allowing for a valid comparison of treatment effects.

From the adjustment formula:

$$\mathbb{E}[Y(T=1) - Y(T=0)] = \mathbb{E}_X[ \mathbb{E}[Y|_{T=1, X}] - \mathbb{E}[Y|_{T=0, X}] ]$$

Assuming both $X$ and $Y$ are discrete, we can rewrite the adjustment formula as

$$\begin{aligned}
ATE & = \mathbb{E}[Y(T=1) - Y(T=0)] \\
    & = \sum_x P(X=x) \left[ \sum_y y P(Y=y|_{T=1, X=x}) - \sum_y y P(Y=y|_{T=0, X=x}) \right]
\end{aligned}$$

based on the Bayes Rule, we can rewrite the above as:

$$\begin{aligned}
ATE & = \mathbb{E}[Y(T=1) - Y(T=0)] \\
    & = \sum_x P(X=x) \left[ \sum_y y \frac{P(Y=y|_{T=1, X=x})}{P(T=1|_{X=x}) P(X=x)} - \sum_y y \frac{P(Y=y|_{T=0, X=x})}{P(T=0|_{X=x}) P(X=x)} \right]
\end{aligned}$$

Without the **positivity assumption**, the denominator in causal calculations would become zero, making it impossible to compute the effect.

A positivity violation occurs when a specific subgroup consistently receives only the treatment or only the control.
Because we never observe the alternative condition for that group, we cannot estimate the causal effect in that context.

> The **positivity assumption** (also called **overlap** or **common support**) simply ensures that the distribution of covariates in the treatment group overlaps with those in the control group.

#### 2.5.1.4 No Interference

The **no interference assumption** states that a treatment applied to one unit does not affect the outcome of others.
Mathematically, a unit's outcome $Y_i$ depends only on its own treatment $T_i$, regardless of the treatment assigned to any other units ($T_1, \dots, T_n$):

$$Y_i (T_1, \dots, T_i, \dots, T_n) = Y_i T_i $$

In the cancer example, one patient receiving chemotherapy has no impact on the health outcomes of any other patients.

#### 2.5.1.5 Consistency

The **consistency assumption** states that if a unit receives treatment $T=t$, the observed outcome $Y$ must equal the potential outcome $Y(t)$.

This implies there are no hidden versions of the treatment; if a patient receives chemotherapy ($T=1$), the outcome cannot change across different settings if all other conditions remain the same.
If the treatment is defined as $T=1$, it must lead to the same potential outcome every time that specific unit receives it.

#### 2.5.1.6 Stable Unit-Treatment Value Assumption (SUTVA)

The **Stable Unit Treatment Value Assumption (SUTVA)** requires that one individual's treatment assignment does not affect another's outcome.
It combines two key ideas: **consistency** and **no interference**.

If SUTVA fails, an individual's potential outcome would depend not only on their own treatment but on the specific combination of treatments given to everyone else in the study.


## 2.6: Causality Inference: Logical Flow

A **causal estimand** is the specific effect or quantity we aim to measure, defined using potential outcomes.
It identifies the relationship we want to understand, such as:

* **Average Treatment Effect (ATE):** The effect on the entire population.
* **Average Treatment Effect on the Treated (ATT):** The effect specifically on those who received the treatment.
* **Conditional Average Treatment Effect (CATE):** The effect within a specific subgroup.

> Causal estimands are defined by potential outcomes, which cannot be directly observed in real-world data because we only ever see one outcome for each individual.

Therefore, the statistical estimand comes into play.
A statistical estimand is a quantity derived from the available observational data and does not have a potential outcome.

**Identification** is the process of translating a **causal estimand** (based on unobservable potential outcomes) into a **statistical estimand** (calculated from observable data).

$$\mathbb{E}[Y(T=1) - Y(T=0)] \xrightarrow{\text{Identification}} \mathbb{E}_X \left[ \mathbb{E}(Y |_{T=1, X}) - \mathbb{E}(Y |_{T=0, X}) \right]$$

Once we have the statistical estimand, we use machine learning estimators (like Linear Regression) to calculate the actual values from observational data.
These models estimate the expected outcomes $\mathbb{E}[Y |_{T=1, X}]$ and $\mathbb{E}[Y |_{T=0, X}]$ based on the variables provided.

The complete logical flow, from defining the target causal estimand to calculating the final estimate, is illustrated in the figure below.

The following section demonstrates how causal and statistical assumptions are applied to transform a **causal estimand** into a **statistical estimand**.

$$ATE = \mathbb{E}[ Y(T=1) - Y(T=0) ]$$

Using the no inference assumption: $ATE = \mathbb{E}[ Y(T=1) ] - \mathbb{E}[ Y(T=0) ]$.

Using the linearity of expectations: $ATE = \mathbb{E}_X \left[ \mathbb{E}[Y(T=1) |_X] - \mathbb{E}[Y(T=0) |_X] \right]$.

Using the law of iterated expectations: $ATE = \mathbb{E}_X \left[ \mathbb{E}[Y(T=1) |_{T=1, X}] - \mathbb{E}[Y(T=0) |_{T=0, X}] \right]$.

Using the unconfoundedness and positivity assumption: $ATE = \mathbb{E}_X \left[ \mathbb{E}[Y |_{T=1, X}] - \mathbb{E}[Y |_{T=0, X}] \right]$ (consistency assumption).

This raises two key questions:

1. **Identification**: How do we convert causal estimands into statistical ones, and which tools facilitate this?
2. **Estimation**: Once we have a statistical estimand, how do we calculate the final estimate, and which tools should be used?

### 2.7.4: Regression Analysis with Treatment and Pre-Test Scores

To refine the estimates, we added pre-test scores as an input variable alongside the treatment indicator.
Separate models were fitted for each grade, assuming a constant treatment effect across all pre-test levels (no interaction).

The following figure illustrates the estimated treatment effect by grade, showing the gap between regression lines for treated and control groups relative to pre-treatment scores.
Generally, the treated groups' lines sit slightly higher than the control groups'.


````python
x_ranges = np.array([[5, 40, 40, 40], [25, 125, 125, 125]])

fig, axs = plt.subplots(nrows=2, ncols=2, figsize=(10, 6), dpi=200)
for j, grade in enumerate(sorted(post["grade"].unique())):
    row_index = j // 2
    col_index = j % 2

    is_grade = electric["grade"] == j + 1
    is_control = electric["treatment"] == 0
    is_treated = electric["treatment"] == 1

    x = np.concatenate(
        [
            electric.loc[is_grade & is_treated, "pre_test"].values,
            electric.loc[is_grade & is_control, "pre_test"].values,
        ]
    )
    y = np.concatenate(
        [
            electric.loc[is_grade & is_treated, "post_test"].values,
            electric.loc[is_grade & is_control, "post_test"].values,
        ]
    )
    t = np.repeat(
        [1, 0], [np.sum(is_grade & is_treated), np.sum(is_grade & is_control)]
    )

    axs[row_index, col_index].scatter(
        electric.loc[is_grade & is_control, "pre_test"].values,
        electric.loc[is_grade & is_control, "post_test"].values,
        marker="o",
        s=15,
        label="Control scores",
    )
    axs[row_index, col_index].scatter(
        electric.loc[is_grade & is_treated, "pre_test"].values,
        electric.loc[is_grade & is_treated, "post_test"].values,
        marker="s",
        s=15,
        label="Treated scores",
    )

    lm_1 = LinearRegression()
    lm_1.fit(np.vstack([x, t]).T, y)
    preds = lm_1.intercept_ + lm_1.coef_[0] * x + lm_1.coef_[1] * t

    axs[row_index, col_index].plot(x[t == 0], preds[t == 0], lw=0.5)
    axs[row_index, col_index].plot(x[t == 1], preds[t == 1], lw=0.5)
    axs[row_index, col_index].set_ylim(bottom=40, top=125)

    axs[row_index, col_index].plot([0, 125], [0, 125], "k--", lw=0.5)

    axs[row_index, col_index].set_xlabel("pre-test, $x_i$")
    axs[row_index, col_index].set_ylabel("post-test, $y_i$")
    axs[row_index, col_index].set_title("grade " + str(j + 1))
    axs[row_index, col_index].legend()

fig.suptitle("Pre-test and post-test scores for Electric Company experiment")
plt.tight_layout()
````

### 2.7.5: Interactions of Treatment Effect with Pre-Treatment Inputs

By including an interaction between pre-test scores and the treatment, the model allows the treatment to affect both the intercept and the slope of the regression.
The following figure illustrates this updated model.

````python
fig, axs = plt.subplots(nrows=2, ncols=2, figsize=(10, 6), dpi=150)

x_ranges = np.array([[5, 40, 40, 40], [25, 125, 125, 125]])
for j in range(4):
    row_index = j // 2
    col_index = j % 2
    ok = electric["grade"] == j + 1
    x = np.concatenate(
        [
            electric.loc[ok & (electric["treatment"] == 1), "pre_test"].values,
            electric.loc[ok & (electric["treatment"] == 0), "pre_test"].values,
        ]
    )
    y = np.concatenate(
        [
            electric.loc[ok & (electric["treatment"] == 1), "post_test"].values,
            electric.loc[ok & (electric["treatment"] == 0), "post_test"].values,
        ]
    )
    t = np.repeat(
        [1, 0],
        [
            np.sum(ok & (electric["treatment"] == 1)),
            np.sum(ok & (electric["treatment"] == 0)),
        ],
    )
    # axs[row_index, col_index].plot(x_ranges[:, j], [40, 125], 'k--', lw=0.5) #commented because this is adding another line
    axs[row_index, col_index].set_xlabel("pre-test, $x_i$")
    axs[row_index, col_index].set_ylabel("post-test, $y_i$")
    axs[row_index, col_index].set_title("grade " + str(j + 1))
    axs[row_index, col_index].scatter(
        electric.loc[ok & (electric["treatment"] == 0), "pre_test"].values,
        electric.loc[ok & (electric["treatment"] == 0), "post_test"].values,
        marker="o",
        s=15,
    )
    axs[row_index, col_index].scatter(
        electric.loc[ok & (electric["treatment"] == 1), "pre_test"].values,
        electric.loc[ok & (electric["treatment"] == 1), "post_test"].values,
        marker="s",
        s=15,
    )
    lm_2 = LinearRegression()
    lm_2.fit(np.vstack([x, t, x * t]).T, y)
    preds = (
        lm_2.intercept_ + lm_2.coef_[0] * x + lm_2.coef_[1] * t + lm_2.coef_[2] * t * x
    )
    axs[row_index, col_index].plot(x[t == 0], preds[t == 0], lw=0.5)
    axs[row_index, col_index].plot(x[t == 1], preds[t == 1], lw=0.5)
    axs[row_index, col_index].set_ylim(bottom=40, top=125)
    axs[row_index, col_index].plot([0, 125], [0, 125], "k--", lw=0.5)

fig.suptitle("Pre-test and post-test scores for Electric Company experiment")
plt.tight_layout()
````

For Grade 4, the initial estimated treatment effect was **3.7** (SE = **1.8**).
After adding pre-test scores, the estimate refined to **1.7** with a significantly lower standard error of **0.7** (increased precision and reduced bias).

```
            coefficient  standard error
(Intercept) 110           1.3
treatment   3.7           1.8
            coefficient  standard error
(Intercept) 42           4.3
treatment   1.7          0.7
pre-test    0.7           0
```

Adding an interaction between treatment and pre-test scores complicates interpretation, as the treatment effect $\beta_{treat} + \beta_{interact} \cdot \text{pre-test}$ now depends on the pre-test value.
In Grade 4, where pre-test scores range from approximately 80 to 120, the estimated effect varies from 1.3 to 2.1 respectively.
To distinguish this score-based variation from actual statistical uncertainty, the following figure superimposes random simulation draws to provide a visual quantify of the estimation's precision.

In this figure, the dark line represents the estimated treatment effect as a function of pre-test scores, derived from the difference between the two regression lines in the previous figure.
For Grade 4, the calculated treatment effect is **1.7** with a standard error of **0.7**—matching the results from the non-interaction model.

### 2.8.4: Question:Understanding Effects of Treatment Using Causal Analysis

(Adapted from Hill and Gelman, Chapter 9, Question 4)

The table below describes a hypothetical experiment on $2400$ persons.
Each row of the table specifies a category of person, as defined by his or her pre-treatment predictor $x$, treatment indicator $T$, and potential outcomes $y^0$, $y^1$.
(For simplicity, we assume unrealistically that all the people in this experiment fit into these eight categories.)

In this study, we define two treatment groups: $T=0$ refers to subjects who do not receive aspirin, while $T=1$ refers to those who do receive aspirin.
Moreover, we consider the covariate $x$, which takes on the value $0$ if the subject is male and $1$ if the subject is female.

| Category | #persons in category | $x$ | $T$ | $y^0$ | $y^1$ |
|:--------:|:--------------------:|:---:|:---:|:-----:|:-----:|
|    1     |         300          |  0  |  0  |   4   |   6   |
|    2     |         300          |  1  |  0  |   4   |   6   |
|    3     |         500          |  0  |  1  |   4   |   6   |
|    4     |         500          |  1  |  1  |   4   |   6   |
|    5     |         200          |  0  |  0  |  10   |  12   |
|    6     |         200          |  1  |  0  |  10   |  12   |
|    7     |         200          |  0  |  1  |  10   |  12   |
|    8     |         200          |  1  |  1  |  10   |  12   |

In making the table we are assuming omniscience, so that we know both $y^0$ and $y=1$ for all observations.
But the (nonomniscient) investigator would only observe $x$, $T$, and $y^T$ for each unit.
(For example, a person in category 1 would have $x=0$, $T=0$, $y=4$, and a person in category 3 would have $x=0$, $T=1$, $y=6$.)

**(a) What is the average treatment effect in this population of 2400 persons?**

$$\hat{y}(0) = \frac{300 \times 4 + 300 \times 4 + 500 \times  4 + 500 \times  4 + 200 \times 10 + 200 \times 10 + 200 \times 10}{300 + 300 + 500 + 500 + 200 + 200 + 200 + 200} = 6$$

$$\hat{y}(1) = \frac{300 \times  6 + 300 \times  6 + 500 \times  6 + 500 \times  6 + 200 \times  12 + 200 \times  12 + 200 \times  12 + 200 \times  12}{300 + 300 + 500 + 500 + 200 + 200 + 200 + 200} = 8$$

_Thus, the average treatment effect $\hat{y}(0) - \hat{y}(1) = 6 - 8 = -2$._

**(b) Is it plausible to believe that these data came from a randomised experiment? Defend your answer.**

_In the event that the data was obtained through a randomised experiment, the expectation is that the observed covariates would be equally distributed between the experimental groups._
_In the present study, $500$ out of $1000$ ($50$) subjects in the $T=0$ group have $x=0$, while $700$ out of $1400$ ($50$) subjects in the $T=1$ group have $x=0$._
_Hence, the data appears to have plausibly originated from a randomised experiment._

**(c) Another population quantity is the mean of $Y$ for those who received the treatment minus the mean of $Y$ for those who did not. What is the relation between this quantity and the average treatment effect?**

_The average for those that received $T=0$ can be given by: $= \frac{300 \times 4 + 300 \times 4 + 200 \times 10 + 200 \times 10}{300 + 300 + 200 + 200} = 6.4$._

_The average for those that received  can be given by: $= \frac{500 /times 6 + 350 \times 6 + 200 \times 12 + 200 \times 12}{500 + 500 + 200 + 200} = 7.7$._

_The treatment effect using these averages is $7.7 - 6.4 = 1.3$._

_The relationship between this population quantity and the average treatment effect can be explained by the fact that this population quantity relies solely on the observed responses, whereas the average treatment effect takes into account both the observed and unobserved responses._
_As a result, the treatment effect obtained from this quantity may differ from that obtained from the average treatment effect._


**(d) For these data, is it plausible to believe that treatment assignment is strongly ignorable given the covariate $x$? Defend your answer.**

_In case of ignorability, then $P(T|_{Y(T=0) = y^0, Y(T=1) = y^1, X=x}) = P(T=0|_{X=x})$._

_Let us consider a couple of cases to check._

_First, let us consider $Y(T=0) = 10$, $Y(T=1) = 12$, $X=0$, then $P(T|_{Y(T=0) = 10, Y(T=1) = 12, X=0}) = \frac{200}{200 + 200} = 0.5$ and $P(T=0|_{X=0})= \frac{300+200}{300 + 500 + 200 + 200} = \frac{5}{12}$._

_Thus, we can see that $P(T |_{Y(T=0) = y^0, Y(T=1) = y^1, X=x}) != P(T=0 |_{X=x})$._

_Next, let us consider $Y(T=0) = 4$, $Y(T=1) = 4$, $X=0$, then $P(T|_{Y(T=0) = 10, Y(T=1) = 12, X=0}) = \frac{300}{300 + 500} = \frac{3}{8}$ and $P(T=0 |_{X=0}) = \frac{300 + 200}{300 + 500 + 200 + 200} = \fraq{5}{12}$._

_Thus, we can see that $P(T |_{Y(T=0) = y^0, Y(T=1) = y^1, X=x} != P(T=0 |_{X=x})$._

_This suggests that the treatment assignment is non-ignorable, as given the covariate, it is not conditionally independent from the potential outcomes._


### 2.8.5: Question Causal Modeling and Assumptions

(Adapted from Hill and Gelman, Chapter 9, Question 6)

**You are consulting for a researcher who has performed a randomised trial where the Treatment was a series of 26 weekly therapy sessions, the control was no therapy, and the outcome was self-report of emotional state one year later. However, most people in the treatment group did not attend every therapy session. In fact there was a good deal of variation in the number of therapy sessions actually attended. The researcher is concerned that her results represent “watered down” estimates because of this variation and suggests adding in another predictor to the model: number of therapy sessions attended. What would you advise her?**

_I would advise against simply adding the number of sessions as a predictor, as this would likely introduce **selection bias**._
_The number of sessions attended is a post-treatment variable influenced by the participants' own choices and initial emotional states, which breaks the benefit of your initial randomisation._

_By conditioning on attendance, you are no longer comparing groups that are identical on average, potentially making the therapy appear more or less effective than it truly is._
_Instead, you should first report the **Intent-to-Treat (ITT)** effect, which maintains the integrity of the randomised groups and provides a realistic estimate of the program's impact._
_If you want to understand the effect of actually attending the sessions, you should use **Instrumental Variable (IV)** analysis to estimate the effect of the treatment "as received."_


### 2.8.6: Question: Understanding the Causal Effect with Different Treatment Effects

Assume that linear regression is appropriate for the regression of an outcome, $y$, on treatment indicator, $T$, and a single confounding covariate, $x$.
Sketch hypothetical data (plotting $y$ versus $x$) and regression lines (for treatment and control group) that represent each of the following situations:

_**(a) No treatment effect**_

_In this case, the regression lines for both the control and treatment groups will be parallel or overlapping with similar slope._

````python
import matplotlib.pyplot as plt
import numpy as np
from sklearn.linear_model import LinearRegression

np.random.seed(0)

t = np.random.binomial(1, 0.5, 100)
x = np.random.normal(0, 1, 100)
y = 2 + 3 * x + np.random.normal(0, 1, 100)

plt.scatter(x[t == 0], y[t == 0], label="Control")
plt.scatter(x[t == 1], y[t == 1], label="Treatment")
plt.legend()
plt.xlabel("x")
plt.ylabel("y")

# Fit regression lines
lr_t0 = LinearRegression().fit(x[t == 0].reshape(-1, 1), y[t == 0].reshape(-1, 1))

lr_t1 = LinearRegression().fit(x[t == 1].reshape(-1, 1), y[t == 1].reshape(-1, 1))

# Plot regression lines
plt.plot(x, lr_t0.predict(x.reshape(-1, 1)))
plt.plot(x, lr_t1.predict(x.reshape(-1, 1)))
````


_**(b) Constant treatment effect**_

_In this case, the regression lines for both the control and treatment groups will be parallel to each other with different y-intercepts._

````python
import matplotlib.pyplot as plt
import numpy as np

np.random.seed(0)

t = np.random.binomial(1, 0.5, 100)
x = np.random.normal(0, 1, 100)
y = 2 + 3 * x + 5 * t + np.random.normal(0, 1, 100)

plt.scatter(x[t == 0], y[t == 0], label="Control")
plt.scatter(x[t == 1], y[t == 1], label="Treatment")
plt.legend()
plt.xlabel("x")
plt.ylabel("y")

# Fit regression lines
lr_t0 = LinearRegression().fit(x[t == 0].reshape(-1, 1), y[t == 0].reshape(-1, 1))

lr_t1 = LinearRegression().fit(x[t == 1].reshape(-1, 1), y[t == 1].reshape(-1, 1))

# Plot regression lines
plt.plot(x, lr_t0.predict(x.reshape(-1, 1)))
plt.plot(x, lr_t1.predict(x.reshape(-1, 1)))
````


_**(c) Treatment effect increasing with x**_

_In this case, the regression lines for the control and treatment groups will have different slopes and different y-intercepts._


````python
import matplotlib.pyplot as plt
import numpy as np

np.random.seed(0)

t = np.random.binomial(1, 0.5, 100)
x = np.random.normal(0, 1, 100)
y = 2 + 3 * x + 5 * x * t + np.random.normal(0, 1, 100)

plt.scatter(x[t == 0], y[t == 0], label="Control")
plt.scatter(x[t == 1], y[t == 1], label="Treatment")
plt.legend()
plt.xlabel("x")
plt.ylabel("y")

# Fit regression lines
lr_t0 = LinearRegression().fit(x[t == 0].reshape(-1, 1), y[t == 0].reshape(-1, 1))

lr_t1 = LinearRegression().fit(x[t == 1].reshape(-1, 1), y[t == 1].reshape(-1, 1))

# Plot regression lines
plt.plot(x, lr_t0.predict(x.reshape(-1, 1)))
plt.plot(x, lr_t1.predict(x.reshape(-1, 1)))
````
