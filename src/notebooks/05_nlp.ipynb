# 5: NLP

Causal inference has many ML applications.
This chapter covers causal methods for text data, which is high-dimensional, unstructured, and complex due to word dependencies.


## 5.1: Causal Concepts in Text Data


### 5.1.1: Roles of Text in Causal Inference

In causal inference, text can serve as treatment, outcome, confounder, or mediator—depending on the research question and variable relationships (Weld et al. 2022; Keith et al. 2020).

* **Text as treatment:** Text can act as a treatment when specific linguistic features (e.g., word choice) are manipulated.
  For example, how does positive or negative language in advertising affect buying choices? (Egami et al., 2022).
* **Text as outcome:** ext can be an outcome, e.g., measuring how a social media algorithm change affects user content.
* **Text as confounder:** Text can confound analysis if it influences both treatment (e.g., review visibility) and outcome (e.g., sales).
  For example, review sentiment may affect both how prominently a review appears and whether customers buy the product.
* **Text as mediator:** Text acts as a mediator when it is an intermediate step between treatment and outcome.
  For example, in political campaigns, message framing (e.g., positive/negative) explains how messages affect voter behaviour.


### 5.1.2: Definitions

This section reviews key terms from Section 2.2.1 as applied to text data.

* **Explanatory variables:** they are independent or predictor variables that influence or explain an outcome.
  In text analysis, they include textual features (e.g., words, topics), linguistic traits (e.g., sentiment, readability), and context (e.g., author, date, platform).
  These model treatment effects on outcomes.
* **Outcome variables:** also known as dependent variables, they are the results being studied.
  In text analysis, these can be quantitative (e.g., sentiment scores) or qualitative (e.g., theme presence), and are central to causal analysis.
* **Unobserved variables:** they consist in unmeasured factors influencing relationships between explanatory and outcome variables.
  In text data, these include latent factors (e.g., author intent) or omitted covariates.
  They cause bias or spurious correlations and are common in text analysis.
* **Unit**: In text analysis, the unit is the smallest text instance studied (e.g., a document, sentence, user post, or message thread).


## 5.2: Encoding Text

This section covers estimating causal effects from text data.

Egami et al. (2022) argue that causal inference is simpler with lower-dimensional text data.
They propose an encoder function $g$ mapping text $\mathbf{T}$ to a causal variable $\mathbf{Z} = g(\mathbf{T})$, requiring $g$ to estimate causal effects from text.
